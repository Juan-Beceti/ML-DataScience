```python
food_2 = food_2[(food_2["Entity"] == "Colombia") | (food_2["Entity"] == "Canada")]
sns.lineplot(food_2, x = "Year", y = "Cocoa beans Production (tonnes)", hue = "Entity")
plt.show()
```

```python
year_tools = food_2.groupby(["Entity","Year"])["Cocoa beans Production (tonnes)"].sum().reset_index(name = "Amount").sort_values(by = "Amount",ascending = False)
year_tools = pd.pivot_table(year_tools, values = "Amount", index = "Year", columns = "Entity")
year_tools = pd.DataFrame(data = year_tools)
year_tools
```

```python
food_colombia = food_2[food_2["Entity"] == "Colombia"]

food_colombia = food_colombia.drop(["Entity"], axis = 1)
sns.heatmap(pd.DataFrame(food_colombia.corr()["Cocoa beans Production (tonnes)"]), annot = True)
plt.show()
```

```python
# Lo háremos para este caso solo para Colombia, luego pasaremos a Canada

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

X = food_colombia["Year"].values.reshape(-1,1)
y = food_colombia["Cocoa beans Production (tonnes)"].values.reshape(-1,1)

X_scaler = MinMaxScaler()
X = X_scaler.fit_transform(X)

y_scaler = MinMaxScaler()
y = y_scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
reg = model.fit(X_train, y_train)

y_train_hat = reg.predict(y_train)
y_test_hat = reg.predict(y_test)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)

X_plot = np.linspace(0,1).reshape(-1,1)
y_plot = reg.predict(X_plot)

plt.plot(X_plot, y_plot, 'r--')

print("Entrenamiento:", r2_score(y_train, y_train_hat))
print("Prueba:", r2_score(y_test, y_test_hat))
```

```python
# Pensemos en ajustar estos datos a una Regresión Polinomial

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 3)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)

reg = LinearRegression().fit(X_train_poly, y_train)

y_train_hat = reg.predict(X_train_poly)
y_test_hat = reg.predict(X_test_poly)

plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)

X_plot = np.linspace(0,1).reshape(-1,1)
X_plot_poly = poly.transform(X_plot)
y_plot = reg.predict(X_plot_poly)

plt.plot(X_plot, y_plot, 'r--')

print("Entrenamiento:", r2_score(y_train, y_train_hat))
print("Prueba:", r2_score(y_test, y_test_hat))
```
